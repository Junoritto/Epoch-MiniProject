```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. `maindata_ver1.2.csv` 불러와서 데이터 프레임 저장

```python
maindata = pd.read_csv('/content/drive/MyDrive/EPOCH/data/maindata_ver1.2.csv')
maindata = pd.DataFrame(maindata)
```

### 2. 컬럼 간 상관관계 분석
2-1. 유의미한 컬럼들과 revenue 간의 상관관계

```python
# 유의미한 컬럼만 추출
main_col = maindata[['budget', 'popularity', 'runtime','vote_average', 'vote_count', 'release_year', 'release_month', 'release_day', 'director_score', 'actor_score', 'revenue']]
maincol_corr = main_col.corr()[['revenue']].drop('revenue')

# 직사각형 상관관계 히트맵 시각화
plt.figure(figsize=(6, 8))
sns.heatmap(maincol_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation of Main Column with Revenue')
plt.show()




# # 시각화
# import matplotlib.pyplot as plt
# import seaborn as sns

# plt.figure(figsize=(12, 8))
# sns.heatmap(maindata_corr, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 5})
# plt.title('Main Data Correlation')
# plt.show
```

2-2. 장르 컬럼과 revenue 컬럼 간의 상관관계

```python
# 장르 컬럼만 추출
genre_col = maindata[['War', 'Thriller',	'Music',	'Fantasy',	'TV Movie',	'Crime',	'Action',	'Documentary',	'Western',	'History',	'Family',	'Mystery',	'Romance',	'Animation',	'Drama',	'Comedy',	'Foreign',	'Adventure',	'Science Fiction',	'Horror', 'revenue']]
genre_corr = genre_col.corr()[['revenue']].drop('revenue')

# 직사각형 상관관계 히트맵 시각화
plt.figure(figsize=(6, 8))
sns.heatmap(genre_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation of Genre Column with Revenue')
plt.show()
```

### 3. 변수에 따른 revenue 변화 시각화
3-1. `vote_count`

```python
vote_count_ascend = maindata['vote_count'].sort_values(ascending=True)

plt.plot(vote_count_ascend, maindata['revenue'], 'o-', markersize=1)
plt.ylabel('Revenue')
plt.title('Revenue Based on Several Variables')
plt.legend(['vote_account', 'director_score', 'actor_score'])
plt.show
```
